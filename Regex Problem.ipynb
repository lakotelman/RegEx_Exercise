```python
import re
file = open('./names.txt', encoding='utf-8')
data = file.read()
file.close()
```

```python
print(data)
```

```python
information = re.findall(r'''
    ([\w+ \w+]*,\s[-\w+ ]+)
     \s+
    ([-\d\w]+@[-.\w\d]+)? 
     \s+
    (\(?\d{3}\)?\s?-?\d{3}-\d{4})?
    \s+ 
    ([\w+ ]*\w+,\s[\w+ ]+)?
    \s*
    (@[-.\d\w]+)?
''', data, re.X)

information
```

```python
persons = []
for person in information: 
    person_dict = { 
        "name": person[0],
        "email":  person[1],
        "phone": person[2],
        "job title": person[3],
        "twitter": person[4]
    }
    persons.append(person_dict)
```

```python
for p in persons:
    print(f"""Name: {p['name']}
Email: {p['email']}
Phone: {p['phone']}
Job Title: {p['job title']}
Twitter Handle: {p["twitter"]}\n""")
```
